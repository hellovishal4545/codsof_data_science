train  model using linear regression as it is coorelated with only one variable TV

```python
from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train = train_test_split(data[['TV']],data[['Sales']],test_size= 0.3,random_state =0 )
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
res= model.predict(X_test)
print(res)
```
